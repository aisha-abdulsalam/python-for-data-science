```python
from bs4 import BeautifulSoup as bs
import pandas as pd 
import requests

url = 'https://www.jumia.com.ng/mlp-stay-connected-deals/smartphones/'
page = requests.get(url)
soup = bs(page.content, 'html5lib')
#print(soup.prettify())
table = soup.find('div', attrs={'class':'-paxs row _no-g _4cl-3cm-shs'})
#print(table.prettify())
articles = table.findAll('article', attrs={'class': 'prd _fb col c-prd'})
#print(articles[0].prettify())
#items = []
df = pd.DataFrame(columns=['name', 'discounted-price N', 'discount N', 'rating'])
for phone in articles:
    item = {}
    item['name'] = phone.find('h3', attrs={'class': "name"}).text
    item['discounted-price N'] = phone.find('div', attrs={'class': "prc"}).text.replace("N","")
    item['original_price N'] = phone.find('div', attrs={'class': "old"}).text.replace("N","")
    item['discount '] = phone.find('div', attrs={'class': "bdg _dsct _sm"}).text
    item['rating'] = phone.find('div', attrs={'class': "stars _s"})
    df = df.append(item, ignore_index = True)

#df.head
df.to_csv('jumia_phone_data.csv')
print('The End')
```
