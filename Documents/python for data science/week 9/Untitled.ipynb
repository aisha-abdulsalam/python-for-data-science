```python
# a. Use pandas package to fetch the data from the link above
import pandas as pd
url = "https://raw.githubusercontent.com/dphi-official/Datasets/master/IT_Salary_Survey_EU_18-20/Survey_2019.csv"
df = pd.read_csv(url)

# b. Plot the histogram of the distribution of all the variables
import matplotlib.pyplot as plt
df.hist(figsize=(15,15))
plt.show()

# c. Plot the heatmap of the correlation matrix of all the variables
import seaborn as sns
corr = df.corr()
sns.heatmap(corr, annot=True, cmap="Blues")
plt.show()

# d. Using Pandas dataframe search, create a new dataframe and perform (a) & (b) on the new dataframe
# For example, let's create a new dataframe with only the rows where Gender is Female
new_df = df[df["Gender"] == "Female"]

# Perform (a) & (b) on the new dataframe
new_df.hist(figsize=(15,15))
plt.show()
```
